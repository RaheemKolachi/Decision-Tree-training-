```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pickle as pk
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/omairaasim/machine_learning/master/project_15_decision_tree_classifier/iphone_purchase_records.csv")
```

```python
x=data.iloc[:,1:3]
y=data.iloc[:,-1:]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.1)
```

```python
dt=DecisionTreeClassifier(criterion="entropy", max_depth=4)
dt.fit(xtrain,ytrain)
```

```python
pred=dt.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,pred)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(20,20))
_=tree.plot_tree(dt,feature_names=x.columns,filled=True)
```
